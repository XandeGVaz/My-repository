## **1-Imports Necessários**

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
from scipy.interpolate import interp1d
import math
```

## **2-Definição da função $f(x)$ a ser integrada**


```python
def f(x):
    return(4/(1 + (x**2)))
```

## **3-Definição da função que calcula $\int_a^b f(x) \,dx$**
*(para um dado número de subintervalos)*

### a)Regra do ponto médio

```python
def intg_pm(a,b,N,f):
    h=(b-a)/N ##calcula tamanho das partições
    som=0 ##somatório da fórmula

    x=np.zeros(N+1)
    x[0]=a

    for k in range(1,N+1):
        x[k]=a+h*k ## valores de x a cada partição h
    
    for k in range (1,N+1):##cálculo do somatório
        xb=(x[k-1]+x[k])/2
        som=som+f(xb) 

    return (h*som)
```

### b)Regra do trapézio

```python
def intg_t(a,b,N,f):
    h=(b-a)/N ##calcula tamanho das partições
    som=0 ##somatório da fórmula

    x=np.zeros(N+1)
    x[0]=a

    for k in range(1,N+1):
        x[k]=a+h*k ## valores de x a cada partição h
    
    for k in range (1,N+1):##cálculo do somatório
        som=som+(f(x[k-1]) + f(x[k]))

    return ((h/2)*som)
```

### c)Regra de Simpson

```python
def intg_s(a,b,N,f):
    h=(b-a)/N ##calcula tamanho das partições
    som=0 ##somatório da fórmula

    x=np.zeros(N+1)
    x[0]=a

    for k in range(1,N+1):
        x[k]=a+h*k ## valores de x a cada partição h
    
    for k in range (1,N+1):##cálculo do somatório
        xb=(x[k-1]+x[k])/2
        som=som+f(x[k-1])+4*f(xb)+f(x[k]) 

    return ((h/6)*som)
```

## **4-Obtenção dos erros**

```python
# Vetor se subintervalos
lN=1000
N=np.linspace(1,1000,lN,dtype=int)

#Vetores de erros:
#a)Regra do ponto médio
errors_pm=np.zeros(lN)
for i in range (0,lN):
    errors_pm[i]=abs(intg_pm(0,1,N[i],f)-math.pi)

#b)Regra do trapézio
errors_t=np.zeros(lN)
for i in range (0,lN):
    errors_t[i]=abs(intg_t(0,1,N[i],f)-math.pi)

#c)Regra de Simpson
errors_s=np.zeros(lN)
for i in range (0,lN):
    errors_s[i]=abs(intg_s(0,1,N[i],f)-math.pi)
```

## **5-Plot de resultados**

*a)Plot para regra do ponto médio*

```python
plt.loglog(N,errors_pm,'-r')
plt.grid()
plt.legend(['Pela regra do ponto médio'])
plt.xlabel('N[número de subintervalos]')
plt.ylabel('erro')
plt.title('Erro em função do númerod e subintervalos')
```

*b)Plot para regra do trapézio*

```python
plt.loglog(N,errors_t,'-r')
plt.grid()
plt.legend(['Pela regra do trapézio'])
plt.xlabel('N[número de subintervalos]')
plt.ylabel('erro')
plt.title('Erro em função do númerod e subintervalos')
```

```python
plt.loglog(N,errors_s,'-r')
plt.grid()
plt.legend(['Pela regra de Simpson'])
plt.xlabel('N[número de subintervalos]')
plt.ylabel('erro')
plt.title('Erro em função do númerod e subintervalos')
```
