```python
import numpy as np
import matplotlib.pyplot as plt

n=10*7 ##Referente ao n√∫mero de sequ√™ncias da fun√ß√£o Xn=a*[Xn-1]*[1-Xn-1]
rng=40000 ##Valor usado para que sejam considerados um maior valor de n√∫meros decimais
          ##    como valores de a
red=0.0001 ## escala usada nas fun√ß√µes para transformar os n√∫meros de 0 a 40000
           ##em n√∫meros de 0 a 4 com v√°rios decimais 

## Procedimento que forma e retorna um vetor com os resultados da fun√ß√£o Xn=a*[Xn-1]*[1-Xn-1]
##para um valor em a, sendo que cada elemento do vetor guardar√° o valor para a uam sequ√™ncia
##de n repeti√ß√µes 

def make_list(a,N):
    x=[0.1]
    for i in range (0,N):
        x.append(a*x[i]*(1-x[i]))
    return x

x,y=[],[]

##  Fun√ß√£o que:
##  1:guarda em um vetor x os valores de a considerados
##  2:faz um vetor em que cada elemento √© tamb√©m um vetor[lista] que guarda
##se diferenciando apenas cada um no valor do termo a considerado

for i in range (0,rng+1):
    x.append(i*red)
    y.append(make_list(i*red,n))

plt.plot(x,y, ls='',marker=',')
plt.grid()
plt.xlabel('Valores de "a"')
plt.ylabel('Possiveis valores da sequencia')
plt.title('Grafico de bifurcacoes')
plt.show()
```
