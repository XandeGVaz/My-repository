```python
import numpy as np
import matplotlib.pyplot as plt
import time

##  Função que pega o fator a como referência para guardar
##em um vetor os resultados para a função Xa=a*[Xa-1]*[1-Xa-1]
##executada 5000 vezes e guardando no vetor o resultado para
##cada execução

def make_list(a):
    x=[0.1]
    for i in range (0,4999): ## 4999 pois se não o vetor terá 5001 resultados
                             ##em função do append
        x.append(a*x[i]*(1-x[i]))
    return x

##  Retorna o tempo de calculo da media dos elementos do vetor x
def tcalc_med(x):
    ti=time.time()
    tam=len(x)
    s=0
    i=0
    while(i<tam):
        s=s+x[i]
        i=i+1
    R=s/tam
    tf= time.time() - ti
    return tf
    
##  Retorna o tempo de calculo da variancia dos elementos do vetor x
def tcalc_var(x):
    ti=time.time()
    tam=len(x)
    s=0
    i=0
    m=calc_med(x)
    while(i<tam):
        s=s+(x[i]-m)**2
        i=i+1
    R=s/(tam-1)
    tf= time.time() - ti
    return (tf)

## Retorna a media dos elementos do vetor x
def calc_med(x):
    tam=len(x)
    s=0
    i=0
    while(i<tam):
        s=s+x[i]
        i=i+1
    return (s/tam)

## Retorna a variancia dos elementos do vetor x
def calc_var(x):
    ti=time.time()
    tam=len(x)
    s=0
    i=0
    m=calc_med(x)
    while(i<tam):
        s=s+(x[i]-m)**2
        i=i+1
    R=s/(tam-1)
    return R
```

## Gráficos de média

```python
##  Vetor dos valores de entrada refentes ao eixo x
##feito dessa forma para conter os valores decimais do termo a[0, 0.1 , 0.2...4.0]
x=np.linspace(1,40,40)*(0.1)

## Vetor do eixo y que conterá os tempos necessários para calcular a media
##para diferentes valores do termo a
y=np.array([])

for i in range(1,41):
    y=np.append(y,tcalc_med(make_list(i*(0.1))))

plt.plot(x,y)
plt.grid()
plt.xlabel('Valor do termo "a"')
plt.ylabel('t[s]')
plt.title('Tempo das medias em funcao de "a"')
plt.show()
```

## Gráficos de  Variância

```python
##  Vetor dos valores de entrada refentes ao eixo x
##feito dessa forma para conter os valores decimais do terma a[0, 0.1 , 0.2...4.0]
x=np.linspace(1,40,40)*(0.1)

## Vetor do eixo y que conterá os tempos necessários para calcular a variancia
##para diferentes valores do termo a
y=np.array([])

for i in range(1,41):
    y=np.append(y,tcalc_var(make_list(i*(0.1))))

plt.plot(x,y)
plt.grid()
plt.xlabel('Valor do termo "a"')
plt.ylabel('t[s]')
plt.title('Tempo das variancias em funcao de "a"')
plt.show()
```
