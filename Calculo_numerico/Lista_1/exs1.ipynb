# Primeira Parte do Exercício

```python
import numpy as np
import random
import time
import matplotlib.pyplot as plt
 
##Função 1 que faz a combinação linear de dois vetores e retorna o tempo
##de operação

def func1(n):
    ti=time.time()
    A=np.random.rand(n)
    B=np.random.rand(n)
    i=random.random()
    j=random.random()
    C=i*A+j*B
    tf=time.time() -ti
    return tf

k=10**4
q=10**8

##Vetor do eixo x do grafico
x=np.linspace(k, q, 5)
##Vetor que conterá os valores de processamento
##com cada entrada
y=np.array([])

for u in range(4,9):
    y=np.append(y,func1(10**u))


plt.plot(x,y)
plt.grid()
plt.xlabel('entrada')
plt.ylabel('t[s]')
plt.show()

plt.loglog(x,y)
plt.grid()
plt.xlabel('entrada')
plt.ylabel('t[s]')
plt.show()


```

```python
##Função que multiplica uma matriz de dimensão n
## m vezes
def func2(n,m):
    ti=time.time()
    M=np.random.rand(n,n)
    for u in range(0,m):
        M=M@M
    
    tf=time.time()-ti
    return tf

##Vetor referente ao eixo x que guarda o numero de entradas
X=np.array([])
X=np.linspace(1*(10**3),4*(10**3),4)

##Vetores referentes ao eixo y que guardarao os tempos de 
##processamento para cada entrada
y1,y2,y3=np.array([]),np.array([]),np.array([])

for u in range(1,5):
    y1=np.append(y1,func2(u*(10*3),2))
    y2=np.append(y2,func2(u*(10*3),3))
    y3=np.append(y3,func2(u*(10*3),4))

plt.plot(X,y1,'r',X,y2,'b',X,y3,'g')
plt.grid()
plt.xlabel('entrada')
plt.ylabel('t[s]')
plt.legend(['Função 1','Função 2','Função 3'])
plt.show()



```
